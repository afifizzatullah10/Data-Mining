```python
import sqlite3

# from https://dev.mysql.com/doc/sakila/en/

file_path = '/content/drive/My Drive/Data Mining/data week 1/sakila.db'
conn = sqlite3.connect(file_path)

data = pd.read_sql('''
    SELECT
        rental.rental_id, rental.rental_date, rental.return_date,
        customer.last_name AS customer_lastname,
        store.store_id,
        city.city AS rental_store_city,
        film.title AS film_title, film.rental_duration AS film_rental_duration,
        film.rental_rate AS film_rental_rate, film.replacement_cost AS film_replacement_cost,
        film.rating AS film_rating
    FROM rental
    INNER JOIN customer ON rental.customer_id == customer.customer_id
    INNER JOIN inventory ON rental.inventory_id == inventory.inventory_id
    INNER JOIN store ON inventory.store_id == store.store_id
    INNER JOIN address ON store.address_id == address.address_id
    INNER JOIN city ON address.city_id == city.city_id
    INNER JOIN film ON inventory.film_id == film.film_id
    ;
''', conn, index_col='rental_id', parse_dates=['rental_date', 'return_date'])

data.head()
```

## Visualize the relationship between salary estimate and company rating

```python
data[data["Rating"].isna()]
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
data['Rating'] = imputer.fit_transform(data[['Rating']])

data[data["Rating"].isna()]
```

```python
data[data["Salary Estimate"].isna()]
```

```python
data['Salary Estimate'].value_counts()
```

```python
pattern = r'\$(\d+)K-\$(\d+)K'

data[['min_salary', 'max_salary']] = data['Salary Estimate'].str.extract(pattern).astype(int)

data
```

```python
data['avg_salary'] = (data['min_salary'] + data['max_salary']) / 2

data.head()
```

```python
data['Rating'].describe()
```

```python
# Count values above 5 in the 'Rating' column
count_above_5 = (data['Rating'] > 5).sum()

# Print the result
print(f"Number of values above 5 in 'Rating': {count_above_5}")
```

```python
data['Rating'] = np.clip(data['Rating'], None, 5)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['avg_salary'], data['Rating'])  # Switched x and y
plt.xlabel("Average Salary")  # Updated x-axis label
plt.ylabel("Rating")  # Updated y-axis label
plt.title("Relationship between Average Salary and Rating")
plt.grid(True)
plt.show()
```

```python
fig = px.scatter(
    data,
    x='avg_salary',
    y='Rating',
    title='Relationship between Average Salary and Rating',
    labels={'avg_salary': 'Average Salary', 'Rating': 'Rating'}
)

fig.update_traces(marker=dict(size=10))  # Adjust marker size if needed
fig.update_layout(width=1000, height=600)  # Set plot dimensions

fig.show()
```

```python
x = data["avg_salary"]
y = data["Rating"]

# Scatter plot
plt.scatter(x, y, alpha=0.6)

# Add a trendline
z = np.polyfit(x, y, 1)  # Fit a linear trendline
p = np.poly1d(z)
plt.plot(x, p(x), "r--", label="Trendline")  # Plot the trendline

plt.xlabel("Average Salary")
plt.ylabel("Rating")
plt.title("Relationship between Average Salary and Company Rating")
plt.legend()
plt.show()
```

## Visualize the number of job postings that require Python, Hadoop, AWS, or Tableau

```python
data['python'] = data['Job Description'].str.lower().str.contains('python').astype(int)
data['hadoop'] = data['Job Description'].str.lower().str.contains('hadoop').astype(int)
data['aws'] = data['Job Description'].str.lower().str.contains('aws').astype(int)
data['tableau'] = data['Job Description'].str.lower().str.contains('tableau').astype(int)

data.head()
```

```python
# Calculate the sum of each column (representing the count of 1s)
column_sums = data[['python', 'aws', 'hadoop', 'tableau']].sum()

# Create the bar graph
plt.figure(figsize=(8, 6))
plt.bar(column_sums.index, column_sums.values)
plt.xlabel("Skills")
plt.ylabel("Number of Job Postings")
plt.title("Demand for Skills in Data Science Jobs")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
fig = px.bar(
    x=column_sums.index,
    y=column_sums.values,
    title='Demand for Skills in Data Science Jobs',
    labels={'x': 'Skills', 'y': 'Number of Job Postings'}
)

fig.update_layout(
    xaxis_tickangle=-45,
    xaxis={'tickmode': 'linear'},  # Ensure all ticks are shown
    autosize=False,
    width=800,
    height=600
)

fig.show()
```

# Visualization

## Line Graphs

```python
stock = yf.Ticker("AAPL")
data = stock.history(period="3mo")

data
```

```python
AAPL_data = yf.Ticker("AAPL").history(period="3mo")
AMZN_data = yf.Ticker("AMZN").history(period="3mo")
META_data = yf.Ticker("META").history(period="3mo")
NFLX_data = yf.Ticker("NFLX").history(period="3mo")
GOOG_data = yf.Ticker("GOOG").history(period="3mo")

AAPL_data["Ticker"] = "AAPL"
AMZN_data["Ticker"] = "AMZN"
META_data["Ticker"] = "META"
NFLX_data["Ticker"] = "NFLX"
GOOG_data["Ticker"] = "GOOG"

stock_data = pd.concat([AAPL_data, AMZN_data, META_data, NFLX_data, GOOG_data])

stock_data
```

```python
tickers = ['AAPL', 'AMZN', 'META', 'NFLX', 'GOOG']

plt.figure(figsize=(12, 6))  # Set figure size

for ticker in tickers:
    subset = stock_data[stock_data['Ticker'] == ticker]
    plt.plot(subset.index, subset['Close'], label=ticker)  # Plot each stock

plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Closing Stock Prices (3 Months)")
plt.legend()  # Show legend
plt.grid(True)  # Add gridlines
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
# Create the Plotly line chart
fig = px.line(
    stock_data,
    x=stock_data.index,
    y='Close',
    color='Ticker',
    title='Closing Stock Prices (3 Months)',
    labels={'x': 'Date', 'y': 'Closing Price'}
)

fig.show()
```

## Bar Graphs

```python
import matplotlib.ticker as mticker

# Group data by ticker and sum the volume
volume_by_ticker = stock_data.groupby('Ticker')['Volume'].sum()

# Create the bar graph
plt.figure(figsize=(10, 6))
plt.bar(volume_by_ticker.index, volume_by_ticker.values / 1e9)  # Divide by 1e9 for billions
plt.xlabel("Stock Ticker")
plt.ylabel("Total Volume (in billions)")  # Update y-axis label
plt.title("Total Trading Volume by Stock")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Format y-axis ticks to display billions
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: '{:.1f}'.format(x)))

plt.tight_layout()
plt.show()
```
